```python
import os
import sys
from pathlib import Path
import site
import ctypes

# 1. ÂØªÊâæ site-packages
venv_root = Path(sys.prefix)
site_packages = venv_root / "Lib" / "site-packages"
print(f"üè• ÂêØÂä®ËØäÊñ≠Á®ãÂ∫èÔºåÊâ´Êèè: {site_packages}\n")

# 2. ÂÖ≥ÈîÆÊñá‰ª∂Ê∏ÖÂçï (Êñá‰ª∂Âêç -> ÊâÄÂú®ÁöÑÂÖ∏ÂûãÂåÖÂêç)
critical_files = {
    "sycl.dll": "intel_cmplr_lib_rt (ÁºñËØëÂô®ËøêË°åÊó∂)",
    "pi_level_zero.dll": "intel_cmplr_lib_ur (GPUÊé•Âè£)", 
    "mkl_core.dll": "mkl (Êï∞Â≠¶Â∫ì)",
    "intel-ext-pt-gpu.dll": "IPEX (PyTorchÊâ©Â±ïÊ†∏ÂøÉ)"
}

found_paths = {}

# 3. ÂÖ®ÁõòÊâ´ÊèèÂØªÊâæÊñá‰ª∂‰ΩçÁΩÆ
print("üîç Ê≠£Âú®ÂÆö‰ΩçÂÖ≥ÈîÆ DLL Êñá‰ª∂...")
for root, dirs, files in os.walk(site_packages):
    for filename in critical_files.keys():
        if filename in files:
            full_path = Path(root) / filename
            found_paths[filename] = full_path
            
            # Á´ãÂç≥Ê≥®ÂÜåËØ•Ë∑ØÂæÑ
            folder = str(full_path.parent)
            if folder not in os.environ['PATH']:
                os.environ['PATH'] = folder + ';' + os.environ['PATH']
                try:
                    os.add_dll_directory(folder)
                except:
                    pass

# 4. ÈÄê‰∏™Â∞ùËØïÂ∫ïÂ±ÇÂä†ËΩΩ (ÊâæÂá∫Ë∞ÅÊòØÂÜÖÈ¨º)
print("\nüíâ Ê≠£Âú®ËøõË°åÂ∫ïÂ±ÇÂä†ËΩΩÊµãËØï...")
all_passed = True

# Êåâ‰æùËµñÈ°∫Â∫èÂä†ËΩΩÔºöÂÖà RuntimeÔºåÊúÄÂêé IPEX
load_order = ["sycl.dll", "pi_level_zero.dll", "mkl_core.dll", "intel-ext-pt-gpu.dll"]

for dll_name in load_order:
    if dll_name not in found_paths:
        print(f"‚ùå [‰∏•ÈáçÁº∫Â§±] Êâæ‰∏çÂà∞Êñá‰ª∂: {dll_name}")
        print(f"   -> ËØ∑Á°ÆËÆ§ {critical_files[dll_name]} ÊòØÂê¶ÂÆâË£ÖÊàêÂäü")
        all_passed = False
        continue

    path = str(found_paths[dll_name])
    try:
        # Â∞ùËØïÁõ¥Êé•Âä†ËΩΩ DLL
        ctypes.CDLL(path)
        print(f"‚úÖ [Âä†ËΩΩÊàêÂäü] {dll_name}")
    except OSError as e:
        all_passed = False
        print(f"‚ùå [Âä†ËΩΩÂ§±Ë¥•] {dll_name}")
        print(f"   -> Ë∑ØÂæÑ: {path}")
        print(f"   -> ÈîôËØØ‰ª£Á†Å: {e}")
        if "126" in str(e):
            print("   üí° ÂàÜÊûê: Êñá‰ª∂Â≠òÂú®‰ΩÜÊó†Ê≥ïÂä†ËΩΩ„ÄÇÂéüÂõ†Ôºö")
            print("      1. Áº∫Â∞ë VC++ Redistributable (ËØ∑ÂéªÂæÆËΩØÂÆòÁΩë‰∏ãËΩΩÂÆâË£Ö)")
            print("      2. Áº∫Â∞ëÂÆÉ‰æùËµñÁöÑÊõ¥Â∫ïÂ±ÇÊñá‰ª∂ (Â¶Ç Level Zero È©±Âä®)")

if not all_passed:
    print("\nüõë ËØäÊñ≠ÁªìÊùüÔºöÂ∫ïÂ±Ç‰æùËµñÊúâ‰∏•ÈáçÈóÆÈ¢òÔºåÊó†Ê≥ïÁªßÁª≠ÂØºÂÖ• Torch„ÄÇ")
    print("ËØ∑Ê†πÊçÆ‰∏äÊñπÁ∫¢Âèâ ‚ùå ÁöÑÊèêÁ§∫‰øÆÂ§çÁéØÂ¢É„ÄÇ")
else:
    print("\nüü¢ ËØäÊñ≠ÁªìÊùüÔºöÂ∫ïÂ±ÇÂ∫ìÂÖ®ÈÉ®ÂÅ•Â∫∑ÔºÅÊ≠£Âú®Â∞ùËØïÊúÄÁªàÂØºÂÖ•...")
    
    # 5. ÊúÄÁªàÂØºÂÖ•
    try:
        import torch
        import intel_extension_for_pytorch as ipex
        print("\nüéâüéâüéâ IPEX ÂØºÂÖ•ÊàêÂäüÔºÅ")
        if torch.xpu.is_available():
            print(f"üöÄ ÊòæÂç°Â∑≤Â∞±Áª™: {torch.xpu.get_device_name(0)}")
        else:
            print("‚ö†Ô∏è ÂØºÂÖ•ÊàêÂäü‰ΩÜÊú™Ê£ÄÊµãÂà∞ÊòæÂç° (ÂèØËÉΩÊòØÈ©±Âä®ÈóÆÈ¢ò)")
    except Exception as e:
        print(f"‚ùå Python Â±ÇÂØºÂÖ•Â§±Ë¥•: {e}")
```

```python
import os
import sys
import shutil
from pathlib import Path
import site

# 1. ÂÆö‰πâÁéØÂ¢ÉÊ†πÁõÆÂΩï
venv_root = Path(sys.prefix)
site_packages = venv_root / "Lib" / "site-packages"

# ÁõÆÊ†áÔºöIPEX ÁöÑËÄÅÂ∑¢
dest_dir = site_packages / "intel_extension_for_pytorch" / "bin"

# 2. ÂÆö‰πâÂøÖÈ°ªÊâæÂà∞Âπ∂Â§çÂà∂ÁöÑ "ÈÄöÁºâÂêçÂçï"
wanted_dlls = [
    "sycl.dll",          # Ê†∏ÂøÉËøêË°åÊó∂
    "sycl6.dll",         # ÂèäÂÖ∂Âèò‰Ωì
    "pi_level_zero.dll", # GPU Êé•Âè£
    "mkl_core.dll",      # MKL Ê†∏ÂøÉ
    "mkl_intel_thread.dll",
    "mkl_sequential.dll",
    "libiomp5md.dll",    # OpenMP (ÊúâÊó∂ÂÄôÂè´Ëøô‰∏™)
    "iomp5md.dll"        # OpenMP (ÊúâÊó∂ÂÄôÂè´Ëøô‰∏™)
]

print(f"üïµÔ∏è‚Äç‚ôÇÔ∏è ÂêØÂä®ÂÖ®Â¢ÉÁåéÊùÄÊ®°Âºè...")
print(f"üìç ÊêúÁ¥¢ËåÉÂõ¥: {venv_root}")
print(f"üéØ ËÅöÂêàÁõÆÊ†á: {dest_dir}\n")

if not dest_dir.exists():
    print("‚ùå ÈîôËØØÔºöIPEX ÁõÆÂΩï‰∏çÂ≠òÂú®ÔºåËØ∑ÂÖàÂÆâË£Ö intel-extension-for-pytorch")
else:
    copied_count = 0
    # 3. ÈÅçÂéÜÊï¥‰∏™ËôöÊãüÁéØÂ¢É (‰∏çÊ≠¢ site-packages)
    for root, dirs, files in os.walk(venv_root):
        # ‰ºòÂåñÔºöË∑≥ËøáÊòæÁÑ∂Êó†ÂÖ≥ÁöÑÊñá‰ª∂Â§πÔºåÂä†ÈÄüÊêúÁ¥¢
        if "__pycache__" in root or "doc" in root:
            continue
            
        for filename in files:
            if filename in wanted_dlls:
                source_path = Path(root) / filename
                target_path = dest_dir / filename
                
                # Â¶ÇÊûúÁõÆÊ†áÊñá‰ª∂Â§πÈáåÊ≤°ÊúâÔºåÊàñËÄÖÊ∫êÊñá‰ª∂Êõ¥Êñ∞ÔºåÂàôÂ§çÂà∂
                if not target_path.exists() or target_path.stat().st_size != source_path.stat().st_size:
                    try:
                        shutil.copy2(source_path, target_path)
                        print(f"‚úÖ ÊçïËé∑Âπ∂Â§çÂà∂: {filename}")
                        print(f"   Ê∫ê: {source_path}")
                        copied_count += 1
                    except Exception as e:
                        print(f"   ‚ùå Â§çÂà∂Â§±Ë¥• {filename}: {e}")

    if copied_count == 0:
        print("\nüíÄ Ëá¥ÂëΩÈîôËØØÔºöÊêúÈÅç‰∫ÜÊï¥‰∏™ÁéØÂ¢ÉÈÉΩÊ≤°ÊâæÂà∞ sycl.dll Êàñ mkl_core.dll„ÄÇ")
        print("ÁªìËÆ∫Ôºö‰Ω†ÁöÑËøêË°åÂ∫ìÁ°ÆÂÆûÊ≤°ÊúâÂÆâË£ÖÊàêÂäü„ÄÇËØ∑ÊâßË°å‰∏ã‰∏ÄÊ≠•„ÄêÂº∫Âà∂ÈáçË£Ö„Äë„ÄÇ")
    else:
        print(f"\nüéâ ÊàêÂäüËÅöÂêà‰∫Ü {copied_count} ‰∏™ÂÖ≥ÈîÆ DLL Êñá‰ª∂ÔºÅ")
        print("‰æùËµñÈìæÂ∑≤Áâ©ÁêÜ‰øÆÂ§ç„ÄÇ")
```

```python
import os
import sys
import shutil
from pathlib import Path

# 1. ÈîÅÂÆö Python ÁúüÊ≠£Êä•ÈîôÁöÑÈÇ£‰∏™ÁõÆÂΩï (Ê†πÊçÆ‰Ω†ÁöÑ Traceback Á°¨ÁºñÁ†Å)
venv_root = Path(sys.prefix)
true_site_packages = venv_root / "Lib" / "site-packages"
target_ipex_bin = true_site_packages / "intel_extension_for_pytorch" / "bin"

print(f"üéØ ‰øÆÊ≠£ÂêéÁöÑÁõÆÊ†á: {target_ipex_bin}")

# 2. ÂÆö‰πâÊâÄÊúâÂèØËÉΩÁöÑ‚ÄúDLL ‰ªìÂ∫ì‚Äù
# ÂåÖÂê´ÂàöÊâçÊîæÈîôÁöÑ‰ΩçÁΩÆÔºå‰ª•Âèä Library/bin
sources = [
    venv_root / "intel_extension_for_pytorch" / "bin", # ÂàöÊâçËÑöÊú¨ÊîæÈîôÁöÑÂú∞Êñπ
    venv_root / "Library" / "bin",                     # Anaconda È£éÊ†ºÁöÑÂ∫ìÁõÆÂΩï
    venv_root / "bin",
    true_site_packages / "intel_cmplr_lib_rt" / "bin",
    true_site_packages / "mkl" / "bin"
]

# 3. ÂºÄÂßãÊê¨Ëøê
if not target_ipex_bin.exists():
    print("‚ùå ÈîôËØØÔºöÁõÆÊ†áÁõÆÂΩï‰∏çÂ≠òÂú®ÔºåËØ∑Á°ÆËÆ§ IPEX ÂÆâË£Ö‰ΩçÁΩÆ„ÄÇ")
else:
    count = 0
    print("üöÄ ÂºÄÂßãÂêëÁõÆÊ†áÊ±áËÅö DLL...")
    
    for src in sources:
        if src.exists():
            # Êê¨ËøêÊâÄÊúâ DLL
            for file in src.glob("*.dll"):
                try:
                    dest_file = target_ipex_bin / file.name
                    # Â¶ÇÊûúÁõÆÊ†á‰∏çÂ≠òÂú®ÔºåÊàñËÄÖÊ∫êÊñá‰ª∂Êõ¥Â§ß(Êõ¥ÂÆåÊï¥)ÔºåÂàôË¶ÜÁõñ
                    if not dest_file.exists() or dest_file.stat().st_size != file.stat().st_size:
                        shutil.copy2(file, dest_file)
                        count += 1
                        # print(f"   -> Â§çÂà∂: {file.name}")
                except Exception as e:
                    pass
    
    print(f"\n‚úÖ ‰øÆÊ≠£ÂÆåÊàêÔºÅÂ∑≤Â∞Ü {count} ‰∏™ DLL Êñá‰ª∂ÂΩí‰ΩçÂà∞Ê≠£Á°ÆÁõÆÂΩï„ÄÇ")
```

```python
import os
import sys
import glob
from pathlib import Path

# 1. ÂÆö‰πâ‰Ω†ÁöÑ site-packages Ë∑ØÂæÑ
env_root = Path(sys.prefix)
site_packages = env_root / "Lib" / "site-packages"

print(f"Ê≠£Âú®Êâ´Êèè DLL Ë∑ØÂæÑ: {site_packages}")

# 2. Êàë‰ª¨ÈúÄË¶ÅÊâæÁöÑÂÖ≥ÈîÆ DLL Êñá‰ª∂Âêç (Ëøô‰∫õÊòØ IPEX ÁöÑÊ†∏ÂøÉ‰æùËµñ)
# sycl*.dll (SYCLËøêË°åÊó∂), pi_level_zero.dll (Level ZeroÊé•Âè£), libiomp5md.dll (OpenMP), mkl*.dll (Êï∞Â≠¶Â∫ì)
patterns = [
    "**/sycl*.dll",
    "**/pi_level_zero.dll",
    "**/libiomp5md.dll", 
    "**/mkl_core.dll"
]

found_dirs = set()

# 3. Êö¥ÂäõÊâ´ÊèèÔºöÊâæÂà∞Ëøô‰∫õ DLL ÊâÄÂú®ÁöÑÊñá‰ª∂Â§π
for pattern in patterns:
    # ÈÄíÂΩíÊêúÁ¥¢
    for dll_path in site_packages.rglob(pattern):
        parent_dir = str(dll_path.parent)
        if parent_dir not in found_dirs:
            found_dirs.add(parent_dir)

# 4. ÊääÊâæÂà∞ÁöÑÊñá‰ª∂Â§πÂä†ÂÖ•ÁéØÂ¢ÉÂèòÈáèÂíå DLL ÊêúÁ¥¢Ë∑ØÂæÑ
if not found_dirs:
    print("‚ö†Ô∏è Ë≠¶ÂëäÔºöÊ≤°ÊúâÂú® site-packages ÈáåÊâæÂà∞ Intel Áõ∏ÂÖ≥ÁöÑ DLL„ÄÇËØ∑Á°ÆËÆ§ pip install ÊòØÂê¶ÁúüÁöÑÊàêÂäü„ÄÇ")
else:
    print(f"‚úÖ ÊâæÂà∞ {len(found_dirs)} ‰∏™ÂÖ≥ÈîÆ DLL ÁõÆÂΩïÔºåÊ≠£Âú®Ê≥®ÂÖ•...")
    for d in found_dirs:
        print(f"  -> Ê∑ªÂä†: {d}")
        # ÈíàÂØπ Windows ÁöÑ PATH
        os.environ['PATH'] = d + ';' + os.environ['PATH']
        # ÈíàÂØπ Python 3.8+ ÁöÑ DLL ÂÆâÂÖ®Âä†ËΩΩÊú∫Âà∂
        try:
            os.add_dll_directory(d)
        except Exception as e:
            print(f"    (os.add_dll_directory Â§±Ë¥•: {e})")

print("-" * 30)

# 5. Áé∞Âú®Â∞ùËØïÂØºÂÖ•
import torch
try:
    import intel_extension_for_pytorch as ipex
    print(f"\nüéâüéâüéâ Â•áËøπÂèëÁîüÔºÅIPEX ÂØºÂÖ•ÊàêÂäüÔºÅÁâàÊú¨: {ipex.__version__}")
    if torch.xpu.is_available():
        print(f"üöÄ XPU ÊòæÂç°Â∞±Áª™: {torch.xpu.get_device_name(0)}")
    else:
        print("‚ö†Ô∏è ÂØºÂÖ•ÊàêÂäüÔºå‰ΩÜÊú™Ê£ÄÊµãÂà∞ XPU ËÆæÂ§á (ÂèØËÉΩÊòØÈ©±Âä®ÈóÆÈ¢òÔºå‰ΩÜÂ∫ì‰øÆÂ•Ω‰∫Ü)")
except Exception as e:
    print(f"\n‚ùå ‰æùÁÑ∂Êä•Èîô (ËØ∑Êää‰∏ãÈù¢ÁöÑÊä•Èîô‰ø°ÊÅØÂèëÁªôÊàë):")
    print(e)
    
    # È¢ùÂ§ñË∞ÉËØï‰ø°ÊÅØÔºöÁúãÁúã torch ÁâàÊú¨ÂØπ‰∏çÂØπ
    print(f"\nÂΩìÂâç Torch ÁâàÊú¨: {torch.__version__}")
```
